```python
import numpy as np
import matplotlib.pyplot as plt
X=np.array([50,120,180,250,320,300]).reshape(-1,1)#units consumed
y=np.array([500,1000,1500,1000,2500,2500])#electricity bill correspondingly
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg=DecisionTreeRegressor(max_depth=3,random_state=42)
dt_reg.fit(X,y)
```

```python
X_grid=np.linspace(0,400,200).reshape(-1,1)
y_pred=dt_reg.predict(X_grid)

plt.scatter(X,y,color='red')
plt.plot(X_grid,y_pred,color='blue')
plt.xlabel("Units Consumed")
plt.ylabel("Electricty bill")
plt.title("DEcison Tree Regression - Electricity Bill slabs")
plt.show()
```

```python
units=float(input("Enter the number of units consumed: "))
bill=dt_reg.predict([[units]])
print("predicted electricity bill",bill[0])
```

Predict salaries based on Years of Experience using Decision Tree Regression
i/p -> Years of Experience
o/p -> Salary

```python
X=np.array([1,2,3,4,5,7,8,9]).reshape(-1,1)
y=np.array([3,3.5,4,4.2,6,6.5,7,7.2])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
dt_reg=DecisionTreeRegressor(random_state=42)
dt_reg.fit(X_train,y_train)
```

```python
X_grid=np.linspace(1,8,100).reshape(-1,1)
y_pred=dt_reg.predict(X_grid)

plt.scatter(X,y,color='red')
plt.plot(X_grid,y_pred,color='blue')
plt.xlabel("years of experience")
plt.ylabel("Salary in LPA")
plt.title("DEcison Tree Regression - Salary Prediction")
plt.show()
```

```python
dt_controlled=DecisionTreeRegressor(max_depth=3)
dt_controlled.fit(X_train,y_train)
y_pred_controlled=dt_controlled.predict(X_grid)
plt.scatter(X,y,color='red')
plt.plot(X_grid,y_pred_controlled,color='blue')
plt.xlabel("years of experience")
plt.ylabel("Salary in LPA")
plt.title("DEcison Tree Regression with controlled depth - Salary Prediction")
plt.show()
```

```python
dt_deep=DecisionTreeRegressor(max_depth=None)
dt_deep.fit(X_train,y_train)
y_pred_deep=dt_deep.predict(X_grid)

plt.scatter(X,y,color='red')
plt.plot(X_grid,y_pred_deep,color='blue')
plt.xlabel("years of experience")
plt.ylabel("Salary in LPA")
plt.title("DEcison Tree Regression with unlimited depth - Salary Prediction")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score

y_test_pred_deep=dt_deep.predict(X_test)
y_test_pred_controlled=dt_controlled.predict(X_test)

print("Deep Tree Metrics")
print("MSE:",mean_squared_error(y_test,y_test_pred_deep))
print("R2 Score:",r2_score(y_test,y_test_pred_deep))

print("Controlled Depth Tree Metrics")
print("MSE:",mean_squared_error(y_test,y_test_pred_controlled))
print("R2 Score:",r2_score(y_test,y_test_pred_controlled))
```
